# Seperating data types 

```python
def seperate_colnames(dataframe):
    global catcols,numcols
    catcols = []
    numcols= []
    for item in dataframe:
        if item == "UNITID":
            continue
        if dataframe[item].max() ==1:
            catcols.append(item)
        else:
            numcols.append(item)
    print(numcols)
    print()
    print(catcols)
            
```

```python
seperate_colnames(data)
```

```python
def numdf(dataframe):
    global numdata
    for col in dataframe.columns:
        if col not in numcols:
            col = col
            del dataframe[col]
    numdata = dataframe
    return dataframe.head()


def catdf(dataframe):
    global catdata
    for col in dataframe.columns:
        if col not in catcols:
            col = col
            del dataframe[col]
    catdata = dataframe
    return dataframe.head()
```

```python
numdf(data1)
```

```python
catdf(data2)
```
