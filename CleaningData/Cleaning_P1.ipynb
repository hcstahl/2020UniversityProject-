# States

```python
data["STATE"]= data["STATE"].replace(["AL","AR","DE","FL","GA","LA","KY","MD","MI","NC","OK","SC","TN","TX","VA","WV","DC","MS"], "SOUTH")
```

```python
data["STATE"]= data["STATE"].replace(["AL","AZ","CA","CO","HI","ID","MT","NV","NM","OR","UT","WA","WY","AK"], "WEST")
```

```python
data["STATE"]= data["STATE"].replace(["CT","ME","MA","NH","NJ","NY","PA","RI","VT"], "NORTH")
```

```python
data["STATE"]= data["STATE"].replace(["IL","IN","IA","KS","MI","MN","MO","NE","ND","OH","SD","WI"], "MidWest")
```

```python
data["STATE"]= data["STATE"].replace(["VI","FM","PR"], "Other")
```
