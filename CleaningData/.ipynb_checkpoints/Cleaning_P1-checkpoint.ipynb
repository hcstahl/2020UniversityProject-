```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.set_option('max_columns', None)
```

```python
data = pd.read_csv("DATA.csv",index_col = "UNITID")
```

```python
data.rename(columns = {'\nTEACH_CERT':'TEACHER_CERT'}, inplace = True)
data.rename(columns = {'STATE_GRANT':'STATEGRANT'}, inplace = True)
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# fixing 0 on salary column
```

```python
data["TOTAL_SALARY_FT_CONTRACT"].describe()
```

```python
Average = data["TOTAL_SALARY_FT_CONTRACT"].mean()
data["TOTAL_SALARY_FT_CONTRACT"].replace(0,Average,inplace=True)
```

```python
data["TOTAL_SALARY_FT_CONTRACT"].describe()
```

```python
data_set = pd.get_dummies(data, columns=['STATE'])
```

```python
# HLOFFER dummy variable 
data_set = pd.get_dummies(data_set,columns=["HLOFFER"])
```

```python
data_set.rename(columns = {'HLOFFER_5':'Highest_Bachelor'}, inplace = True)
data_set.rename(columns = {'HLOFFER_6':'Highest_PostBachelor'}, inplace = True)
data_set.rename(columns = {'HLOFFER_7':'Highest_Masters'}, inplace = True)
data_set.rename(columns = {'HLOFFER_8':'Highest_PostMasters'}, inplace = True)
data_set.rename(columns = {'HLOFFER_9':'Highest_Doc'}, inplace = True)
```

```python
print(data_set["Highest_Bachelor"].sum())
print(data_set["Highest_PostBachelor"].sum())
print(data_set["Highest_Masters"].sum())
print(data_set["Highest_PostMasters"].sum())
print(data_set["Highest_Doc"].sum())
```

```python
data_set.drop('Highest_PostBachelor',axis=1,inplace=True)
```
